```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.svm import SVC
```

# EXO1

1. Given dataset

```python
class0 = np.array([[1, 9], [5, 5], [1, 1]])
class1 = np.array([[8, 5], [13, 1], [13, 9]])
```

2. Concatenate the two matrices and provide the label vector


```python
X = np.vstack((class0, class1))
y = np.hstack((np.zeros(len(class0)), np.ones(len(class1))))
```

3. Plot the points in a scatter plot


```python
plt.scatter(class0[:, 0], class0[:, 1], color='blue', label='Class 0')
plt.scatter(class1[:, 0], class1[:, 1], color='red', label='Class 1')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Scatter Plot of Data Points')
plt.legend()
plt.grid(True)
plt.show()
```

4. Train a linear SVM classifier with maximum margin

```python
clf = SVC(kernel='linear')
clf.fit(X, y)
```



5.   Plot decision boundary







```python
plt.scatter(class0[:, 0], class0[:, 1], color='blue', label='Class 0')
plt.scatter(class1[:, 0], class1[:, 1], color='red', label='Class 1')
w = clf.coef_[0]
b = clf.intercept_[0]
x0 = np.linspace(0, 14)
decision_boundary = -w[0]/w[1] * x0 - b/w[1]
plt.plot(x0, decision_boundary, 'k-', label='Decision Boundary')
plt.scatter(clf.support_vectors_[:, 0], clf.support_vectors_[:, 1], s=100, facecolors='none', edgecolors='k', label='Support Vectors')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Linear SVM Decision Boundary with Maximum Margin')
plt.legend()
plt.grid(True)
plt.show()
```

# EXO2

1. Given dataset

```python
class0 = np.array([[2, 2], [3, 3], [4, 4], [5, 5], [4, 6], [3, 7], [4, 8], [5, 9], [6, 10]])
class1 = np.array([[6, 2], [7, 3], [8, 4], [9, 5], [8, 6], [7, 7], [7, 8], [7, 9], [8, 10]])
```

2. Plot the points in a scatter plot

```python
plt.scatter(class0[:, 0], class0[:, 1], color='blue', label='Class 0')
plt.scatter(class1[:, 0], class1[:, 1], color='red', label='Class 1')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Scatter Plot of Data Points')
plt.legend()
plt.grid(True)
plt.show()
```

3. Train a linear SVM classifier with a high cost

```python
clf = SVC(kernel='linear', C=1000)
clf.fit(X, y)
```

4. Plot decision boundary


```python
plt.scatter(class0[:, 0], class0[:, 1], color='blue', label='Class 0')
plt.scatter(class1[:, 0], class1[:, 1], color='red', label='Class 1')
w = clf.coef_[0]
a = -w[0] / w[1]
xx = np.linspace(2, 9)
yy = a * xx - (clf.intercept_[0]) / w[1]
plt.plot(xx, yy, 'k-', label='Decision Boundary')
plt.scatter(clf.support_vectors_[:, 0], clf.support_vectors_[:, 1], s=100, facecolors='none', edgecolors='k', label='Support Vectors')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Linear SVM Decision Boundary with High Cost')
plt.legend()
plt.grid(True)
plt.show()
```


6. Calculate the error rate

```python
error_rate = 1 - clf.score(X, y)
print("Error rate of the linear SVM on the training dataset:", error_rate)
```
